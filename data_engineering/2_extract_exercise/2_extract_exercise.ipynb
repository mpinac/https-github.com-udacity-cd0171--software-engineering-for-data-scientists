```python
def print_lines(n, file_name):
    f = open(file_name)
    for i in range(n):
        print(f.readline())
    f.close()

print_lines(1, 'population_data.json')
```

```python
# import the BeautifulSoup library
from bs4 import BeautifulSoup

with open('population_data.xml') as fp:
    soup = BeautifulSoup(fp, 'lxml') #lxml is the parser type
```

```python
# output the first 5 record in the xml file
# this is an example of how to navigate the XML document with BeautifulSoup

i = 0
# use the find_all_method to get all fields in each record
for record in soup.find_all('record'):
    # use the find all method to get all fields in each record
    i += 1
    for r in record.find_all('field'):
        print(r['name'], ':', r.text)
    print()
    if i == 5:
        break
```

```python
xml_dict = {}

# use the find_all_method to get all fields in each record
for record in soup.find_all('record'):
    # use the find all method to get all fields in each record
    for r in record.find_all('field'):
        xml_dict.setdefault(r['name'], []).append(r.text)

population_df = pd.DataFrame(xml_dict)
```
